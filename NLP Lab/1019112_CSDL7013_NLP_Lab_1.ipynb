Q. 2. Remove all spaces, punctuation symbols. Store all words in a list, find the total number of words, print some parts of the list, find top ten words of maximum length and print them.

[*02 Marks*]

```python
# Solultion to Q.2.
# alice.txt
import re
punct = re.compile("[^\w\s]")
a = open("alice.txt", "r")
new_a = "alice_new.txt"
f = a.read()
new_f = re.sub(punct, "", f)
with open(new_a, "w") as fw:
    fw.write(new_f)
lst1 = open("alice_new.txt").read().split()
print("Number of words: ",len(lst1))
print(lst1[2:10])
lst2 = []
for i in range(10):
  res = max(lst1, key = len)
  lst2.append(res)
  lst1.remove(res)
print("Top 10 words of max length: ",lst2)
# holmes.txt
import re
punct = re.compile("[^\w\s]")
b = open("holmes.txt", "r")
new_b = "holmes_new.txt"
f = b.read()
new_f = re.sub(punct, "", f)
with open(new_b, "w") as fw:
    fw.write(new_f)
lst3 = open("holmes_new.txt").read().split()
print("Number of words: ",len(lst3))
print(lst3[2:10])
lst4 = []
for i in range(10):
  res = max(lst3, key = len)
  lst4.append(res)
  lst3.remove(res)
print("Top 10 words of max length: ",lst4)
```

Q. 3. Consider the following stopwords (Add few more if you feel some are left out):
    
    {'a','an','the','of','is','this','that','to'}

Extract all words from the given input excluding the stopwords and put it in the list. Print the count of such words.

[*02 Marks*]

```python
# Solultion to Q.3.
# alice.txt
lst1 = open("/content/alice_new.txt").read().split()
lst1 = [x.lower() for x in lst1]
stopwords = ['a','an','the','of','is','this','that','to','there','they','off','was','when','i','it','and','but','so','in']
lst2 = []
for i in lst1:
  if i not in stopwords:
    lst2.append(i)
print(len(lst2))

# holmes.txt
lst3 = open("/content/holmes_new.txt").read().split()
lst3 = [x.lower() for x in lst3]
stopwords = ['a','an','the','of','is','this','that','to','there','they','off','was','when','i','it','and','but','so','in']
lst4 = []
for i in lst3:
  if i not in stopwords:
    lst4.append(i)
print(len(lst4))
```

Q. 4.  Perform frequency distribution on the list in the previous exercise.

[*02 Marks*]

```python
# Solultion to Q.4.
# alice.txt
freq1 = {}
for item in lst2:
   if item in freq1:
      freq1[item] += 1
   else:
      freq1[item] = 1
print(freq1)
```

```python
# holmes.txt
freq2 = {}
for item in lst4:
   if item in freq2:
      freq2[item] += 1
   else:
      freq2[item] = 1
print(freq2)
```

Q. 5. Plot the frequency distribution graph.

[*02 Marks*]

```python
# alice.txt
import matplotlib.pyplot as plt
plt.figure(figsize=(100, 50))
plt.rc('font',size=35)
plt.bar(list(freq1.keys()),list(freq1.values()),align='center', width=0.8,color="lightblue")
plt.ylabel("Frequency")
plt.xlabel("Words")
plt.xticks(rotation=45)
plt.show()
```

```python
# holmes.txt
import matplotlib.pyplot as plt
plt.figure(figsize=(100, 50))
plt.rc('font',size=35)
plt.bar(list(freq2.keys()),list(freq2.values()),align='center', width=0.8,color="lightpink")
plt.ylabel("Frequency")
plt.xlabel("Words")
plt.xticks(rotation=45)
plt.show()
```
